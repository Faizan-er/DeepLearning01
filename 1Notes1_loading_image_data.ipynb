## Iterating through dictnaries

```python
Student_dict = {
    'Rizwan': 'age= 21',
    'Faizan': 'age= 24',
    'Imran': 'age= 19',
    'Rehan': 'age = 26'
}
```

```python
for a, b  in Student_dict.items():
    print(a)
    print(b)
```

```python
for a in Student_dict.keys():
    print(a)
```

```python
for a in Student_dict.values():
    print(a)
```
